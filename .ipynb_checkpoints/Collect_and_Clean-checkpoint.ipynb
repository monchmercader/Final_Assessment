## Australia Youtube Data section

```python
#Collect all AU data and concatenate into AU dataframe
#Use the path above and add /datasets
#This uses an absolute path and os.path.join to make it OS independent

#For Macs path, replace the path below to your own:
#path = r'//Users/monchmercader/Python/Data_Analytics/Module 1/Final_Assessment/datasets'

#For Windows path, replace the path below to your own:
path = r'C:\Users\monch.mercader\Python\Data_Analytics\Module 1\Final_Assessment\datasets'
all_AU_files = glob.glob(os.path.join(path, "????????_AU_videos.csv"))     
df_from_each_AU_file = (pd.read_csv(f) for f in all_AU_files)
Total_AU_df = pd.concat(df_from_each_AU_file, ignore_index=True)
```

```python
#Checking if the above operation was successful, and how many entries there are
Total_AU_df.info()
```

```python
#Quick check on the top 3 entries
Total_AU_df.head(3)
```

```python
#Showing the columns for the dataframe
Total_AU_df.columns
```

```python
#Filter categories to the most relevant, categoryId: 19 - Travel & Events, 21 - Video blogging, 22 - People & Blogs, 26 - How To & Style, 37 Family)
#Passing into a new filtered dataframe
Total_AU_df_filtered = Total_AU_df.drop(Total_AU_df[(Total_AU_df['categoryId'] != 19) & (Total_AU_df['categoryId'] != 21) 
                                                    & (Total_AU_df['categoryId'] != 22) & (Total_AU_df['categoryId'] != 26) 
                                                    & (Total_AU_df['categoryId'] != 37)].index) 
```

```python
#Checking on how many entries are remaining
Total_AU_df_filtered.info()
```

```python
#Filtering out duplicates: Since the raw CSV scrapes are collected daily, videos may be popular over a period of time. Therefore, it is necessary to filter out the duplicates, and retain the most recent occurance of the video_id.
#This can be done by first finding out how many duplicates a video_id has. 
#The lines below uses the video_id column to pivot and counts the number of instances it appears in the dataframe

dup_df = pd.DataFrame(Total_AU_df_filtered, columns= ['video_id'])

duplicates_df = dup_df.pivot_table(index=['video_id'], aggfunc='size')
print (duplicates_df)
#Copy one of the video IDs below and use it in the following cells to check
```

```python
#This is to check that the drop_duplicates action in the next cells keep the correct entry (highest video)
# Insert your selected video ID from above here: Total_AU_df_filtered.loc[Total_AU_df_filtered['video_id'] == '<INSERT VIDEO ID HERE>']
Total_AU_df_filtered.loc[Total_AU_df_filtered['video_id'] == 'zBZ4skqOm04']
#Take note of the largest view_count number for this video
```

```python
#This uses the drop method to remove the duplicates under the video_id column, and to keep the last instance. 
Final_AU_df = Total_AU_df_filtered.drop_duplicates(subset=['video_id'], keep='last')
```

```python
#Quick check on how many entries remain after filtering duplicates
Final_AU_df.info()
```

```python
Final_AU_df.head(3)
```

```python
#To confirm that the process was successful, use the video ID you selected and confirm that the highest view count is the only entry that remains 
Final_AU_df.loc[Final_AU_df['video_id'] == 'zBZ4skqOm04']
```
